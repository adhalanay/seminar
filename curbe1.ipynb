```python
def frenet_plane(c):
    T=derivative(c,t)
    T=T/norm(T)
    N=[T[1],-T[0]]
    return(T,N)
```

```python
c(t)=[9*cos(t),4*sin(t)]
```

```python
var('t')
(T,N)=frenet_plane(c)
vectors_T=[arrow((c[0](t),c[1](t)),(c[0](t)+2*T[0](t),c[1](t)+2*T[1](t)), color=hue(0.3)) for t in srange(0,2*pi,0.3)]
vectors_N=[arrow((c[0](t),c[1](t)),(c[0](t)+2*N[0](t),c[1](t)+2*N[1](t)),color=hue(0.6)) for t in srange(0,2*pi,0.3)]
backgr=[parametric_plot((c[0],c[1]), (0,2*pi),color=hue(0.7)) for nr in srange(0,2*pi,0.3)]
aT=animate(vectors_T,xmin=-11,xmax=11,ymin=-5.5,ymax=5.5)
aN=animate(vectors_N,xmin=-11,xmax=11,ymin=-5.5,ymax=5.5)
aP=animate(backgr,xmin=-11,xmax=11,ymin=-5.5,ymax=5.5)
(aT+aN+aP).show()
```

```python
reset('u')
reset('s')
reset('t')
reset('x')
def curbura(c):
    var('t')
    g(t)=c(t)
    xp=derivative(g[0])
    yp=derivative(g[1])
    xpp=derivative(xp)
    ypp=derivative(yp)
    rez=(xp*ypp-xpp*yp)/(xp**2+yp**2)**(3/2)
    return rez
```

```python
def c(x):
    return (3*cos(x),2*sin(x))
```
